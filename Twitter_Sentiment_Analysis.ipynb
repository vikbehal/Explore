###Import libraries

```
import tweepy
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```
# My Twitter app Authentication variables
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```
#Authenticate
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

#Get API
api = tweepy.API(auth)

#Get tweets for 'Artificial Intelligence'
tweets = api.search('Artificial Intelligence', count=200)

#Build data frame
data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])
```

```
#Check some of the tweets
display(data.head(10))
```

```
print(tweets[0].created_at)
```

```
import nltk
nltk.download('lexicon')
```

```
sid = SentimentIntensityAnalyzer()

listy = []

for idx, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
  
#Conver to Series
se = pd.Series(listy)

#Add new column to DataFrame
data['polarity'] = se.values
```

```
import nltk
nltk.download('vader_lexicon')
```

```
sid = SentimentIntensityAnalyzer()

listy = []

for idx, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
  
#Conver to Series
se = pd.Series(listy)

#Add new column to DataFrame
data['polarity'] = se.values
```

###Tweaks

```
#Get tweets for 'Artificial Intelligence'
tweets1 = api.search('Artificial Intelligence', count=200, tweet_mode='extended')

#Build data frame
data1 = pd.DataFrame(data=[tweet.full_text for tweet in tweets1], columns=['Tweets'])
```

```
#Check some of the tweets
display(data1.head(10))
```

```
print(tweets1[0].created_at)
```

```
sid = SentimentIntensityAnalyzer()

listy1 = []

for idx, row in data1.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy1.append(ss)
  
#Conver to Series
se1 = pd.Series(listy1)

#Add new column to DataFrame
data1['polarity1'] = se1.values
```
